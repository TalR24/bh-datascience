## Haver database and Python

This exercise will introduce you to the Haver library designed for Python. 

```python
### This is an introduction to the Haver package within Python.
### We will be working with a set of example databases that come 
### with the Haver package and are stored in the root directory.

## Let's import the Haver package into Python
```

```python
import Haver
```

```python
## Check the current path the Haver package will point at

Haver.path() 
```

```python
## Set the path to the example database built into Haver

Haver.path('examples')
```

```python
## Check the contents of the examples database

Haver.databases()
```

```python
## Let's get an overview of what is inside some of the databases
```

```python
Haver.metadata(database='haverd')
```

```python
Haver.metadata(database='haverw')
```

```python
Haver.metadata(database='havermqa')
```

```python
## You can always assign output to a variable. In this case we are 
## assigning a haver metadata frame to the variable hmd_mqa. 
## The result is a pandas DataFrame.

hmd_mqa = Haver.metadata(database='havermqa')
```

```python
## Let's query from one of the example databases

hd_1 = Haver.data(['fxtwb', 'fxtwm', 'fxtwotp'], 'haverd', startdate='2012-01-01')
```

```python

type(hd_1)
```

```python
hd_1
```

```python
## You can pull data from multiple databases

hd_2 = Haver.data(['haverd:fxtwb', 'haverw:farbn'], startdate='2012-01-01')
hd_2 ## Can you determine what happened with the frequency of the data
```

```python
## Earlier we created a variable hmd_mqa which held a Haver metadata object.
## We can use that assigned variable as a query 

hd_mqa = Haver.data(hmd_mqa)
hd_mqa
```

```python
## You can specify frequency in your data request. In this case, the query
## will drop the annual series and only return quarterly series

hd_mqa2=Haver.data(hmd_mqa, frequency='q')
hd_mqa2.iloc[-10:, 0:4]
```

```python
## We can quickly plot the data to get a feel for the series we are intersted in

import matplotlib.pyplot as plt
```

```python
hd_mqa2.loc[:, ['mm', 'ms']].plot() 

## loc gets rows (or columns) with particular labels from the index.
## iloc gets rows (or columns) at particular positions in the index (so it only takes integers).
```

```python
plt.show()
```
